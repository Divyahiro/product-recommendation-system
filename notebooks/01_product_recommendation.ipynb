# Select Required Columns
This step selects the columns required for building the product recommendation system.

```python
recommendation_df = df[["reviews.username", "name", "reviews.rating"]]
recommendation_df.head()
```

# Check for missing Values
This step checks missing values in the selected columns.

```python
recommendation_df.isnull().sum()
```

# Remove Missing Values
This step removes rows containing missing values from the selected columns.

```python
recommendation_df = recommendation_df.dropna()
recommendation_df.head()
```

# Rename Columns
This step renames the columns to simpler names for easier analysis.

```python
recommendation_df.columns = ["user", "product", "rating"]
recommendation_df.head()
```

# Create User-Product Matrix
This step creates a matrix where rows represent users, columns represent products, and values represent ratings.

```python
user_product_matrix = recommendation_df.pivot_table(
    index="user",
    columns="product",
    values="rating"
)

user_product_matrix.head()
```

# Fill Missing Ratings
This step replaces missing ratings with 0 to prepare the data for recommendation.

```python
user_product_matrix = user_product_matrix.fillna(0)

user_product_matrix.head()
```

# Display Matrix Dimensions
This step displays the number of users and products in the recommendation matrix.

```python
user_product_matrix.shape
```

# Calculate User Similarity
This step calculates the similarity between users based on their product ratings.

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_product_matrix)

user_similarity.shape
```

# Convert User Similarity Matrix to DataFrame
This step converts the user similarity matrix into a DataFrame for easier interpretation.

```python
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_product_matrix.index,
    columns=user_product_matrix.index
)

user_similarity_df.head()
```

# Find Similar Users
This step displays the users most similar to a selected user.

```python
sample_user = user_similarity_df.index[0]

user_similarity_df[sample_user].sort_values(ascending=False).head(10)
```

# Find Top Rated Products
This step identifies the highest-rated products in thr dataset.

```python
top_products = (
    recommendation_df.groupby("product")["rating"]
    .mean()
    .sort_values(ascending=False)
)

top_products.head(10)
```

# Recommend Products for a User
This step recommends products based on the ratings of similar users.

```python
sample_user = recommendation_df["user"].iloc[0]

rated_products = recommendation_df[
    recommendation_df["user"] ==sample_user
]["product"]

recommended_products = top_products[~top_products.index.isin(rated_products)]

recommended_products.head(10)
```

# Save Recommendations
This step saves the recommended products to a CSV file.

```python
recommended_products.head(10).to_csv("../models/recommended_products.csv")
```
